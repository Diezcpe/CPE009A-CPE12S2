Prelim Exam Q29

```python
#Input marks for subjects
marks = []
n = int(input("Enter number of subjects: "))

for i in range(n):
    m = int(input(f"Enter marks for subject {i+1}: "))
#f is formatted string literal
    marks.append(m)

total_marks = sum(marks)
percentage = total_marks / n
```

```python
#Function to assign grade based on marks (canvas given)
def get_grade(mark):
    if 90 <= mark <= 100:
        return "A+"
    elif 80 <= mark <= 89:
        return "A"
    elif 70 <= mark <= 79:
        return "B+"
    elif 60 <= mark <= 69:
        return "B"
    elif 50 <= mark <= 59:
        return "C"
    elif 40 <= mark <= 49:
        return "D"
    else:
        return "F"
```

```python
#Performance remarks based on percentage (canvas given)
if percentage >= 90:
    remark = "Outstanding"
elif percentage >= 80:
    remark = "Excellent"
elif percentage >= 70:
    remark = "Very Good"
elif percentage >= 60:
    remark = "Good"
elif percentage >= 40:
    remark = "Satisfactory"
else:
    remark = "Needs Improvement"
```

```python
print("\nSubject Grades:")
for i, m in enumerate(marks):
    print(f"Subject {i+1}: Marks = {m}, Grade = {get_grade(m)}")
#f is formatted string literal

result = "Pass" if all(m >= 40 for m in marks) else "Fail"
#Display grades for each subject
```

```python
#Overall
print("\nFinal Result")
print("Total Marks:", total_marks)
print("Percentage:", percentage)
print("Overall Result:", result)
print("Performance Remark:", remark)
```
