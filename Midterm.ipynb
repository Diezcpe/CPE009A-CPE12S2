The Break Statement

```python
week = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday", "Saturday"]
for x in week:
  if x == "Thursday":
    break
  print(x) #indent
```

```python
week = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday", "Saturday"]
for x in week:
  if x == "Thursday":
    break
print(x) #no indent
```

The Range () Function

```python
for x in range(6):
  print(x)
for x in range(2,6):
  print(x)
```

```python
for v in range(2,6,2): #start at 2 increment 2 then next is 4 stop at 6
  print(v)
```

Nested Loops

```python
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for a in adjective:
  for b in fruits:
    print(a,b)
```

Pyhton While Loop

```python
#Example 5
i = 1
while i < 6:
  print(i)
  i += 1 #i+1 assign operator, print before increment 1 then go to while loop again till i<6 then exit
```

The break statement in While loop

```python
i = 1
while i<6:
  print(i) #first output 1
  if i== 3: #false
    break #stop if the number reach 3
  i += 1
```

The Continue Statement

```python
#Continuous if the statement is true
i = 0
while i < 6:
  i += 1
  if i == 3: #skip three then goes back to while loop
    continue #since its true it will continue but it will not print 3
  print(i)
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print(i)
  continue
```

```python
i = 1
while i < 6:
  print(i) #print before increment
  i += 1
else:
  print("i is no longer less than 6")
```
