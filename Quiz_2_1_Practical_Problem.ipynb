```python
# We want to create a python code for a grading scheme
# Create a function to accept a number of activities
# Compute for the score vs total
# Return the average grade of the student in %
def termCrit():
    inputScore = float(input("Score: "))
    inputTotal = float(input("Total: "))

    if inputScore > inputTotal:
        print("\nInvalid Scores.\n")
        return 0.0

    if inputTotal == 0:
        return 0.0

    grade = (inputScore / inputTotal) * 100
    return grade
#I remove 'for i in range(actNumber)" because it causes loop"

# Create a function for the class standing per term
def termCS():
    print("Hands-on Activities: ")
    termHOA = termCrit()

    print("Quizzes: ")
    termQuiz = termCrit()

    print("Assignments: ")
    termAss = termCrit()

    grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
    return grade
#I also remove "int input(asking how many terms)" because it causes errors

# Create a function to compute the term grade
def termGrade(termName):
    print(f"\nComputing for the {termName} grade.")
    examRaw = float(input("Exam Score: "))
    examTotal = float(input("Exam Total: "))
    examGrade = (examRaw / examTotal) * 100

    CSgrade = termCS()
    grade = (0.5 * examGrade) + (0.5 * CSgrade)
    return grade
#Basically, I just add "termName"

# Create a function to compute for the grades in the semeste
def semesterGrade():
    prelimGrade = termGrade("Prelim")
    midtermGrade = termGrade("Midterm")
    finalGrade = termGrade("Final")

    finalMG = ((1/3) * prelimGrade) + ((2/3) * midtermGrade)
    finalFG = ((1/3) * finalMG) + ((2/3) * finalGrade)

    return finalFG
#I just arrange it in short form and add parenthesis in the computing grades

#The class should add everything else
def convertGrade(finalGrade):
    if finalGrade >= 99.0: return 1.0
    elif finalGrade >= 96.0: return 1.25
    elif finalGrade >= 93.0: return 1.50
    elif finalGrade >= 90.0: return 1.75
    elif finalGrade >= 87.0: return 2.00
    elif finalGrade >= 84.0: return 2.25
    elif finalGrade >= 81.0: return 2.50
    elif finalGrade >= 78.0: return 2.75
    elif finalGrade >= 75.0: return 3.0
    elif finalGrade >= 74.0: return 5.0
 #I also add the remaining grading system base on the canvas

# Driver Code
print("\nFinal Grade Equivalent:", convertGrade(semesterGrade()))
```
