```python
# Tuples
tuplelist = (1, 2, 3, 4, 5)
print(tuplelist[0])
print(tuplelist[1])
print(tuplelist[2])
print(tuplelist[3])
print(tuplelist[4])
print(tuplelist + (1, 2, 3))

#Cause an error
# Attempting to change a tuple element results in a TypeError")

print("Slicing tuplelist:")
print(tuplelist[0:3])  # First three elements
print(tuplelist[2:])   # From index 2 to end
print(tuplelist[:4])   # From start to index 3
print(tuplelist[-3:])  # Last three elements
print(tuplelist[1:4:2])  # Step slicing

print("Printing tuple elements with for loop:")
for num in tuplelist:
    print(num)
```

```python
# Dictionaries
contactinfo = {'id': 1, 'first_name': 'John', 'last_name': 'Doe', 'contact_number': '09060611233'}
print(contactinfo['id'])
print(contactinfo['first_name'])
print(contactinfo['last_name'])
print(contactinfo['contact_number'])
# print(contactinfo['age']) - This would cause KeyError because 'age' doesn't exist
# KeyError would occur for contactinfo['age'] because the key doesn't exist")

# 3. Type the code: for k,v in contactinfo: print(k)
# This will cause an error - need to iterate differently

print("Dictionary keys:")
for k in contactinfo:
    print(k)
print("Dictionary items:")
for k, v in contactinfo.items():
    print(k, v)
contactinfo['id'] = 2
contactinfo['first_name'] = 'Max'
print("Updated contactinfo:")
print(contactinfo)
```

```python
# Supplementary Activity:
# 1. Distance Formula
print("\n--- Distance Formula ---")
coordinates_list = [(1, 1), (2, 3)]
x1, y1 = coordinates_list[0]
x2, y2 = coordinates_list[1]
distance = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
print(f"Coordinates: {coordinates_list}")
print(f"Distance between {coordinates_list[0]} and {coordinates_list[1]}: {distance}")

# 2. Simple Word Filter
print("\n--- Simple Word Filter ---")
text = input("Enter a string: ")
filtered_text = text.replace("stupid", "*" * len("stupid"))
print("Filtered text:", filtered_text)

# 3. Phonebook
print("\n--- Phonebook Program ---")
phonebook = [
    {'id': 1, 'name': 'John Doe', 'contact': '09060611233', 'address': '123 Main St'},
    {'id': 2, 'name': 'Jane Smith', 'contact': '09123456789', 'address': '456 Oak Ave'},
    {'id': 3, 'name': 'Bob Johnson', 'contact': '09987654321', 'address': '789 Pine Rd'}
]

try:
    search_id = int(input("Enter ID to search (1-3): "))
    found = False
    for record in phonebook:
        if record['id'] == search_id:
            print(f"\nRecord found:")
            print(f"Name: {record['name']}")
            print(f"Contact: {record['contact']}")
            print(f"Address: {record['address']}")
            found = True
            break
    if not found:
        print("Record not found")
except ValueError:
    print("Please enter a valid integer ID")
```
