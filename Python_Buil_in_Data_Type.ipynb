Functions

```python
def remainder(n,m):
  while True:
    if n - m < 0: #1st if 10-4 < 0 (False)
                  #2nd if use n as 6 base on else, 6-4 < 0 (False)
                  #3rd if use n as 2 base on second else, 2-4 < 0 (True)
      return n
    else: #Proceed to else, n= 10-4 = 6, (Proceed again to 2nd if)
          #2nd else n= 6-4 = 2
      n = n - m
remainder(10,4)
```
