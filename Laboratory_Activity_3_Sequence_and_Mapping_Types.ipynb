```python
numberlist = [5,4,2,1,3]
len(numberlist)
print(numberlist[0])
print(numberlist[1])
print(numberlist[2])
print(numberlist[3])
print(numberlist[4])
print(numberlist[-1])
print(numberlist[-2])
print(numberlist[-3])
print(numberlist[-4])
print(numberlist[-5])
#print(numberlist[5]) and print(numberlist[-6]) are error
```

```python
itemlist = [1,-2.0,[1,2,3],'Word']
len(itemlist)
print(itemlist [0])
print(itemlist [1])
print(itemlist [2])
print(itemlist [3])
len(itemlist[2])
print(itemlist [2][0])
print(itemlist [2][1])
print(itemlist [2][2])
print(itemlist [-1])
print(itemlist [-2])
print(itemlist [-3])
print(itemlist [-4])
len(itemlist[-2])
print(itemlist[-2][0])
print(itemlist[-2][1])
print(itemlist[-2][2])
print(itemlist[-2][-3])
print(itemlist[-2][-2])
print(itemlist[-2][-1])
#LEN prints the

```

```python
longlist = numberlist + itemlist
len(longlist)
print(longlist [:])
print(longlist[:9])
print(longlist[0:])
print(longlist[1:])
print(longlist[2:])
#Index Slicing with Range
print(longlist[2:5])
print(longlist[5:2])
print(longlist[8:])
print(longlist[9:])
#Index Slicing using Negative Indices
print(longlist[-9:])
print(longlist[-8:])
print(longlist[-8:-7])
print(longlist[-1:])
#Other properties of Index Slicing
print(longlist[10:20])
print(longlist[-7:5])
#Index Slicing with Step parameter
print(longlist[::1])
print(longlist[::2])
print(longlist[1:8:2])
print(longlist[9:1:-1])
print(longlist[-1::1])
print(longlist[-1::-1])
```
